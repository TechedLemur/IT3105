```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from simworld import SimWorld
import random
```

```python
s = SimWorld()
```

```python
lines = np.zeros((300,2,3))
for i in range(300):
    x = [0,s.state.xp2,s.state.x_tip]
    y = [0,s.state.yp2,s.state.y_tip]
    s.do_action(random.randint(0,2))
    # s.do_action(0)
    lines[i,0] = x 
    lines[i,1] = y 
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.set(xlim=(-2.5, 2.5), ylim=(-2.5, 2.5))

line = ax.plot(lines[0, 0], lines[0, 1], 'o-' , color='k', lw=2)[0]

def animate(i):
    line.set_xdata(lines[i, 0])
    line.set_ydata(lines[i, 1])
    # return line

# t = np.linspace(1, 25, 30)

anim = FuncAnimation(
    fig, animate, interval=100)#, frames=len(t)-1)

plt.draw()
plt.show()
```
